```python
# Import os module for file operations
import os

# Define the source and destination directories
source_dir = "/home/bkcs/HDD/Transfer-Learning-Library/examples/domain_adaptation/image_classification/Test/Change_lambda/S2T/"
dest_dir = "/home/bkcs/HDD/Transfer-Learning-Library/examples/domain_adaptation/image_classification/Result/Conf/"

# Define a function to recursively search for conf.pdf files in a given directory
def find_conf_files(dir_path):
    # Loop through all the files and subdirectories in the given directory
    for entry in os.scandir(dir_path):
        # If the entry is a file and its name is conf.pdf
        if entry.is_file() and entry.name == "conf.pdf":
            # Yield the full path of the file
            yield entry.path
        # If the entry is a subdirectory
        elif entry.is_dir():
            # Recursively search for conf.pdf files in the subdirectory
            yield from find_conf_files(entry.path)

# Loop through all the subdirectories in the source directory
for sub_dir in os.listdir(source_dir):
    # Construct the full path of the subdirectory
    sub_dir_path = os.path.join(source_dir, sub_dir)
    # Check if the subdirectory is a directory
    if os.path.isdir(sub_dir_path):
        # Loop through all the conf.pdf files found in the subdirectory
        for conf_file_path in find_conf_files(sub_dir_path):
            # Construct the new file name based on the subdirectory name and the relative path of the conf.pdf file
            new_file_name = f"{sub_dir}_{conf_file_path[len(sub_dir_path)+1:].replace('/', '_')}"
            # Construct the full path of the new file in the destination directory
            new_file_path = os.path.join(dest_dir, new_file_name)
            # Copy and rename the conf.pdf file to the destination directory
            os.system(f"cp {conf_file_path} {new_file_path}")
```

```python
# Import os and re modules for file operations and regular expressions
import os
import re

# Define the source and destination directories
source_dir = "/home/bkcs/HDD/Transfer-Learning-Library/examples/domain_adaptation/image_classification/Result/DAN/percent"
dest_dir = "/home/bkcs/HDD/Transfer-Learning-Library/examples/domain_adaptation/image_classification/Result/Acc_loss"

# Define a function to recursively search for acc_loss.txt files in a given directory
def find_acc_loss_files(dir_path):
    # Loop through all the files and subdirectories in the given directory
    for entry in os.scandir(dir_path):
        # If the entry is a file and its name is acc_loss.txt
        if entry.is_file() and entry.name == "acc_loss.txt":
            # Yield the full path of the file
            yield entry.path
        # If the entry is a subdirectory
        elif entry.is_dir():
            # Recursively search for acc_loss.txt files in the subdirectory
            yield from find_acc_loss_files(entry.path)

# Loop through all the acc_loss.txt files found in the source directory
for acc_loss_file_path in find_acc_loss_files(source_dir):
    # Open the acc_loss.txt file for reading
    with open(acc_loss_file_path, "r") as f:
        # Read the content of the file as a string
        content = f.read()
        # Replace ":" symbol with "," symbol
        content = content.replace(":", ",")
        # Remove "[" and "]" symbols using regular expressions
        content = re.sub(r"\[|\]", "", content)
    # Construct the new file name based on the relative path of the acc_loss.txt file
    new_file_name = f"{acc_loss_file_path[len(source_dir)+1:].replace('/', '_')}"
    # Change .txt to .csv extension
    new_file_name = new_file_name.replace(".txt", ".csv")
    # Construct the full path of the new file in the destination directory
    new_file_path = os.path.join(dest_dir, new_file_name)
    # Open the new file for writing
    with open(new_file_path, "w") as f:
        # Write the modified content to the new file
        f.write(content)
```

```python
# Import the glob and pandas libraries
import glob
import pandas as pd

# Define the folder path and the output file name
folder_path = "/home/bkcs/HDD/Transfer-Learning-Library/examples/domain_adaptation/image_classification/Result/Acc_loss"
output_file = "/home/bkcs/HDD/Transfer-Learning-Library/examples/domain_adaptation/image_classification/Result/Acc_loss_final/"
# Create a list of file names
files = [f for f in os.listdir(folder_path) if f.endswith(".csv")]

# Create an empty dictionary to store the data frames
data_frames = {}

# Loop through the files and read them as data frames
for file in files:
    # Get the key for the data frame from the file name
    key = file.split("use")[0]
    # Read the file as a data frame
    df = pd.read_csv(os.path.join(folder_path, file), header=None)
    postfix = file.split("use")[1]
    postfix = postfix.replace('_visualize_acc_loss.csv','')
    # Get the first column as a series
    col = df.iloc[:, 0]

    # Add the prefix and an underscore to each value
    col = col + postfix

    # Replace the first column with the modified series
    df.iloc[:, 0] = col
    
    # Check if the key already exists in the dictionary
    if key in data_frames:
        # Append the data frame to the existing one
        data_frames[key] = pd.concat([data_frames[key], df], ignore_index=True)
    else:
        # Store the data frame in the dictionary with the key
        data_frames[key] = df

# Loop through the keys and data frames in the dictionary
for key, df in data_frames.items():
    # Create a new file name with the key
    new_file = f"{output_file}{key}.csv"
    # Write the data frame to a new file
    df.to_csv(new_file, index=False)
```
